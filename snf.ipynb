```python
import numpy as np
import pandas as pd

date_cols = [col for col in df.columns if 'DT' in col]

for col in date_cols:
    df[col] = df[col].replace(['nan', 'None', 'NaT'], np.nan)
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
for col in date_cols:
    print(col, df[col].dtype, df[col].notna().sum())
```

```python
# Train Synthesizer


from sdv.metadata import SingleTableMetadata
from sdv.single_table import GaussianCopulaSynthesizer

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=df)

synthesizer = GaussianCopulaSynthesizer(metadata)
synthesizer.fit(df)
```

```python
synthetic_rows = synthesizer.sample(len(df) * 3)
```

```python
# Restore Datetime Columns
for col in date_cols:
    synthetic_rows[col] = pd.to_datetime(synthetic_rows[col], unit='s', errors='coerce')
```

```python
# Fix Claim Structure
max_clm_id = df['CLM_ID'].max()
synthetic_rows['CLM_ID'] = synthetic_rows['CLM_ID'] + max_clm_id + 1
```

```python
max_bene_id = df['BENE_ID'].max()
synthetic_rows['BENE_ID'] = synthetic_rows['BENE_ID'] + max_bene_id + 1
```

```python
# Combine
final_df = pd.concat([df, synthetic_rows], ignore_index=True)

print("Original Shape:", df.shape)
print("Final Shape:", final_df.shape)
```

```python
output_path = "snf_4x_synthetic.csv"

final_df.to_csv(
    output_path,
    index=False,
    encoding="utf-8",
    chunksize=100000  # writes in chunks (memory safe)
)

print("CSV file saved as:", output_path)
```
