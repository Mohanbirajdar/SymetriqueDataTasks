```python
import pandas as pd
import numpy as np

from sdv.metadata import SingleTableMetadata
from sdv.single_table import GaussianCopulaSynthesizer

# Load Inpatient Data

file_path = r"/content/inpatient.csv"

df = pd.read_csv(file_path, sep="|", low_memory=False)

print("Original Shape:", df.shape)
```

```python
# Convert Date Columns Properly
date_columns = [col for col in df.columns if "DT" in col]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
# Engineer Medical Summary Features
if "CLM_FROM_DT" in df.columns and "CLM_THRU_DT" in df.columns:
    df["LENGTH_OF_STAY"] = (df["CLM_THRU_DT"] - df["CLM_FROM_DT"]).dt.days
```

```python
# Number of Diagnosis Codes
diagnosis_cols = [col for col in df.columns if "ICD_DGNS_CD" in col]

df["NUM_DIAGNOSES"] = df[diagnosis_cols].notna().sum(axis=1)
```

```python
# Number of Procedure Codes
procedure_cols = [col for col in df.columns if "ICD_PRCDR_CD" in col]

df["NUM_PROCEDURES"] = df[procedure_cols].notna().sum(axis=1)
```

```python
# Total Procedure Date Count

procedure_date_cols = [col for col in df.columns if "PRCDR_DT" in col]

df["NUM_PROCEDURE_DATES"] = df[procedure_date_cols].notna().sum(axis=1)
```

```python
# Drop Raw High-Cardinality Columns

# We drop:

# Raw diagnosis codes

# Raw procedure codes

# Raw procedure dates

# IDs

# NPIs / Provider identifiers


drop_columns = (
    diagnosis_cols +
    procedure_cols +
    procedure_date_cols +
    ["CLM_ID", "DESYNPUF_ID"]
)

# Drop only columns that actually exist
drop_columns = [col for col in drop_columns if col in df.columns]

model_df = df.drop(columns=drop_columns)

print("After Feature Engineering Shape:", model_df.shape)
```

```python
# Handle Negative or Missing Values

model_df = model_df.fillna(0)

# Ensure no negative financial values
amt_columns = [col for col in model_df.columns if "AMT" in col.upper()]

for col in amt_columns:
    model_df[col] = model_df[col].abs()
```

```python
# Detect Metadata
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(model_df)

# Optional but recommended: save metadata
metadata.save_to_json("intpatient_metadata.json")
```

```python
# rain Gaussian Copula Model

# This is stable for healthcare structured data.
```

```python
synthesizer = GaussianCopulaSynthesizer(metadata)

synthesizer.fit(model_df)
```

```python
# Generate 3× Synthetic Data
num_synthetic = len(model_df) * 3

synthetic_data = synthesizer.sample(num_synthetic)

print("Synthetic Shape:", synthetic_data.shape)
```

```python
# Recreate New IDs
synthetic_data["CLM_ID"] = ["SYN_CLM_" + str(i) for i in range(len(synthetic_data))]
synthetic_data["DESYNPUF_ID"] = ["SYN_BENE_" + str(i) for i in range(len(synthetic_data))]
```

```python
# Recombine Original + Synthetic
final_df = pd.concat([df, synthetic_data], ignore_index=True)

print("Final Dataset Shape:", final_df.shape)
```

```python
# saving file
final_df.to_csv("Inpatient_Claims_4x_Synthetic.csv", index=False)

print("✅ 4x Synthetic Dataset Saved Successfully")
```
